```python
## We have written a helper class in InceptionModule.py to load the model 
## We can use the class to classify some images

model = InceptionModule.InceptionModule()
```

```python
## Lets define a method to classify images
def classify(imagePath):
    # Display the image
    display(Image(imagePath))
    
    # Use the Inception model to classify the image.
    pred = model.classify(image_path=imagePath)
    
    # Print the top 10 classes
    model.print_scores(pred=pred,k=10,only_first_name=True)
```

```python
image_path = 'inception/cropped_panda.jpg'
classify(image_path)
```

```python
classify(imagePath='inception/plane.jpg')
```
